# Import files

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot with Legend and Grid

```python
x = np.arange(1, 11)
y = x * 2

plt.figure(figsize=(6,4))
plt.plot(x, y, label="y = 2x", marker="o")

plt.title("Line Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.legend()
plt.grid(True)

plt.show()
```

# Multiple Lines

```python
y1 = x * 2
y2 = x * 3

plt.plot(x, y1, label="2x")
plt.plot(x, y2, label="3x")

plt.title("Multiple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()

plt.show()
```

# Scatter Plot

```python
x = np.random.randint(1,50,30)
y = np.random.randint(1,50,30)

plt.scatter(x,y)

plt.title("Scatter Plot")
plt.xlabel("X Points")
plt.ylabel("Y Points")

plt.show()
```

# Save Plot

```python
x = np.arange(1,6)
y = [5,7,8,6,9]

plt.plot(x,y, label="Sample Data", marker="o")

plt.title("Saved Plot Example")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()

plt.savefig("sample_plot.png")

plt.show()
```
