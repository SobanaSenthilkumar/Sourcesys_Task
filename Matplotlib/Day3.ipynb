# importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Create sample data

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

sales = [100, 150, 200, 180]
products = ["Laptop", "Phone", "Tablet", "Watch"]

data = np.random.normal(50, 10, 100)
```

# Create Subplots Layout

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))
```

# 1. Line Plot with Annotation

```python
axes[0,0].plot(x, y1, label="sin(x)", color="blue")
axes[0,0].plot(x, y2, label="cos(x)", color="red")

axes[0,0].set_title("Trigonometric Functions")
axes[0,0].set_xlabel("X values")
axes[0,0].set_ylabel("Y values")

axes[0,0].legend()
axes[0,0].grid(True)

# Annotation example
axes[0,0].annotate("Peak", xy=(1.5,1), xytext=(3,1.5),
                   arrowprops=dict(facecolor='black'))
```

# 2. Bar Chart

```python
axes[0,1].bar(products, sales, color="orange")

axes[0,1].set_title("Product Sales")
axes[0,1].set_xlabel("Products")
axes[0,1].set_ylabel("Sales")
```

# 3. Scatter Plot with Colormap

```python
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
colors = np.random.rand(100)

scatter = axes[1,0].scatter(x_scatter, y_scatter,
                             c=colors,
                             cmap="viridis")

axes[1,0].set_title("Scatter Plot with Color Map")

# Add colorbar
fig.colorbar(scatter, ax=axes[1,0])
```

# 4. Histogram

```python
axes[1,1].hist(data, bins=10, color="purple", edgecolor="black")

axes[1,1].set_title("Data Distribution")
axes[1,1].set_xlabel("Values")
axes[1,1].set_ylabel("Frequency")
```

# Combined

```python
import matplotlib.pyplot as plt
import numpy as np

# sample data
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

sales = [100,150,200,180]
products = ["Laptop","Phone","Tablet","Watch"]

data = np.random.normal(50,10,100)

# create subplot layout
fig, axes = plt.subplots(2,2, figsize=(12,8))

# Line plot
axes[0,0].plot(x,y1,label="sin(x)")
axes[0,0].plot(x,y2,label="cos(x)")
axes[0,0].set_title("Trigonometric Functions")
axes[0,0].legend()
axes[0,0].grid(True)

# Bar chart
axes[0,1].bar(products,sales)
axes[0,1].set_title("Product Sales")

# Scatter plot
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

axes[1,0].scatter(x_scatter,y_scatter)
axes[1,0].set_title("Scatter Plot")

# Histogram
axes[1,1].hist(data,bins=10)
axes[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```
