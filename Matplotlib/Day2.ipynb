# Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot
Explanation:

Inputs

X values → days

Y values → temperature

When to use

Trend analysis

Time-series data

```python

days = [1,2,3,4,5]
temperature = [30,32,31,29,35]

plt.plot(days, temperature, marker='o', label="Temperature")

plt.title("Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.legend()
plt.grid(True)

plt.show()
```

# Bar Chart
Explanation:

Inputs

Categories

Values

When to use

Comparing groups

```python
# Bar Chart
# Used to compare categories

products = ["Laptop","Phone","Tablet","Watch"]
sales = [150,200,120,100]

plt.bar(products, sales)

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.show()
```

# Scatter Plot
Explanation:

Inputs

X values

Y values

When to use

Correlation analysis

```python
# Scatter Plot
# Used to show relationship between two variables

study_hours = [1,2,3,4,5,6]
marks = [40,45,50,60,65,70]

plt.scatter(study_hours, marks)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

# Histogram
Explanation:

Inputs

Numerical dataset

Number of bins

When to use

Distribution analysis

```python
# Histogram
# Used to show data distribution

marks = [55,60,62,65,70,72,75,78,80,82,85,88,90]

plt.hist(marks, bins=5)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

# Pie Chart
Explanation:

Inputs

Values

Labels

When to use

Percentage comparison

```python
# Pie Chart
# Used to show percentage contribution

market_share = [40,30,20,10]
companies = ["Company A","Company B","Company C","Company D"]

plt.pie(market_share, labels=companies, autopct="%1.1f%%")

plt.title("Market Share")

plt.show()
```
