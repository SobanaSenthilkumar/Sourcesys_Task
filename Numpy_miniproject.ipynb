# Scenario: E-Commerce Sales Analytics System



## 1️ Introduction to NumPy

```python
import numpy as np
np.random.seed(42)
print('NumPy Version:', np.__version__)
```

## 2️ Creating E-Commerce Dataset

```python
# Number of orders
num_orders = 1000

# Order IDs
order_ids = np.arange(1, num_orders + 1)

# Product prices (₹100 - ₹5000)
prices = np.random.randint(100, 5000, num_orders)

# Quantity (1-5)
quantity = np.random.randint(1, 6, num_orders)

# Discount % (0-30)
discount = np.random.randint(0, 31, num_orders)

# Combine into dataset
dataset = np.column_stack((order_ids, prices, quantity, discount))

print('Dataset Shape:', dataset.shape)
dataset[:5]
```

## 3️ Array Properties

```python
print('Shape:', dataset.shape)
print('Dimensions:', dataset.ndim)
print('Size:', dataset.size)
print('Data Type:', dataset.dtype)
```

## 4️ Revenue Calculation (Mathematical Operations)

```python
total_price = prices * quantity
final_price = total_price - (total_price * discount / 100)

print('Sample Final Prices:', final_price[:5])
```

## 5️ Statistical Functions

```python
print('Average Order Value:', np.mean(final_price))
print('Max Order Value:', np.max(final_price))
print('Min Order Value:', np.min(final_price))
print('Standard Deviation:', np.std(final_price))
print('Total Revenue:', np.sum(final_price))
```

## 6️ Boolean Indexing

```python
high_value_orders = order_ids[final_price > 10000]
print('High Value Orders Count:', high_value_orders.size)
```

## 7️ Reshaping & Axis Operations

```python
reshaped = final_price[:100].reshape(10,10)
print('Row-wise Sum:', np.sum(reshaped, axis=1)[:3])
print('Column-wise Mean:', np.mean(reshaped, axis=0))
```

## 8️ Sorting & Top Orders

```python
top5 = np.sort(final_price)[-5:]
print('Top 5 Orders:', top5)
```

## 9️ Broadcasting

```python
extra_tax = 0.05
price_with_tax = final_price + (final_price * extra_tax)
print('Price with 5% Tax Sample:', price_with_tax[:5])
```

## 10 Linear Algebra Example

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print('Matrix Multiplication:\n', np.dot(A,B))
print('Determinant of A:', np.linalg.det(A))
```

## 11 Saving & Loading Data

```python
np.save('ecommerce_revenue.npy', final_price)
loaded_data = np.load('ecommerce_revenue.npy')
print('Loaded Sample:', loaded_data[:5])
```

## 12 Final Business Insights

```python
high_discount_orders = np.sum(discount > 20)

print('Total Revenue: ₹', np.sum(final_price))
print('Average Revenue per Order: ₹', np.mean(final_price))
print('Orders with >20% Discount:', high_discount_orders)
```
