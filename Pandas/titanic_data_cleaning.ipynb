# Titanic Dataset – Data Cleaning Steps
This notebook demonstrates common data cleaning operations on the Titanic dataset using **Pandas and NumPy**.

```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\VICKY\OneDrive\Desktop\Sourcesys\Pandas\Titanic-Dataset.csv")

# Display first rows
print('Original Dataset')
df.head()
```

## 1. Dataset Information

```python
df.info()
```

## 2. Check Missing Values

```python
df.isnull().sum()
```

## 3. Handle Missing Values

```python
# Fill Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column (many missing values)
df = df.drop(columns=['Cabin'])
```

## 4. Convert Names to Lowercase

```python
df['Name'] = df['Name'].str.lower()
```

## 5. Remove Extra Spaces

```python
df['Name'] = df['Name'].str.strip()
```

## 6. Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

## 7. Encode Categorical Data

```python
# Convert Sex to numeric
df['Sex'] = df['Sex'].map({'male':0, 'female':1})

# One-hot encoding for Embarked
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

## 8. Fix Data Types

```python
df['Age'] = df['Age'].astype(int)
```

## 9. Remove Outliers (Age)

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Age'] >= Q1 - 1.5 * IQR) & (df['Age'] <= Q3 + 1.5 * IQR)]
```

## 10. Normalize Fare

```python
df['Fare'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())
```

## 11. Save Cleaned Dataset

```python
df.to_csv('cleaned_titanic.csv', index=False)

print('Data Cleaning Completed Successfully')
df.head()
```
