# Global Superstore Dataset – Data Cleaning, Preprocessing, Filtering, Sorting, Analysis and Grouping


```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

print("Libraries Loaded Successfully")
```

## Load Dataset

```python
df = pd.read_csv("superstore.csv")

print("Dataset Loaded Successfully")

print("\nFirst 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nColumn Names")
print(df.columns)
```

# BASIC DATA INSPECTION

```python
print("\nDataset Info")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

print("\nChecking Unique Values")
print(df.nunique())
```

## Data Cleaning

```python
print("\nChecking Missing Values")
print(df.isnull().sum())


# Fill missing categorical values
df["Region"] = df["Region"].fillna("Unknown")
df["Category"] = df["Category"].fillna("Unknown")
df["Sub-Category"] = df["Sub-Category"].fillna("Unknown")


# Fill missing numerical values
df["Sales"] = df["Sales"].fillna(df["Sales"].median())
df["Profit"] = df["Profit"].fillna(df["Profit"].median())
df["Discount"] = df["Discount"].fillna(0)


# Remove duplicate rows
df = df.drop_duplicates()

print("\nAfter Removing Duplicates")
print(df.shape)


# Convert column names to lowercase
df.columns = df.columns.str.lower()

# Remove leading/trailing spaces
df.columns = df.columns.str.strip()

print("\nCleaned Column Names")
print(df.columns)
```

## Data Preprocessing

```python
# Convert date columns
df["order date"] = pd.to_datetime(df["order date"])
df["ship date"] = pd.to_datetime(df["ship date"])


# Extract useful features
df["order_year"] = df["order date"].dt.year
df["order_month"] = df["order date"].dt.month
df["order_day"] = df["order date"].dt.day


# Create profit ratio column
df["profit_ratio"] = df["profit"] / df["sales"]


# Convert categorical columns to lowercase
df["category"] = df["category"].str.lower()
df["sub-category"] = df["sub-category"].str.lower()


# Remove whitespace
df["category"] = df["category"].str.strip()
df["sub-category"] = df["sub-category"].str.strip()


print("\nPreprocessed Dataset")
print(df.head())
```

## Filtering

```python
high_sales = df[df["sales"] > 500]

print("\nHigh Sales Orders")
print(high_sales.head())


# Filter furniture category
furniture_sales = df[df["category"] == "furniture"]

print("\nFurniture Category Orders")
print(furniture_sales.head())


# Filter profitable orders
profitable_orders = df[df["profit"] > 0]

print("\nProfitable Orders Count")
print(len(profitable_orders))


# Filter orders from specific region
west_region = df[df["region"] == "West"]

print("\nWest Region Orders")
print(west_region.head())
```

## Sorting

```python
sorted_sales = df.sort_values(by="sales", ascending=False)

print("\nTop Sales Orders")
print(sorted_sales.head())


# Sort by profit
sorted_profit = df.sort_values(by="profit", ascending=False)

print("\nTop Profit Orders")
print(sorted_profit.head())


# Sort by order year and sales
sorted_year_sales = df.sort_values(by=["order_year", "sales"], ascending=[True, False])

print("\nSorted by Year and Sales")
print(sorted_year_sales.head())
```

## Analysis

```python
total_sales = df["sales"].sum()
print("\nTotal Sales:", total_sales)


# Average sales
avg_sales = df["sales"].mean()
print("Average Sales:", avg_sales)


# Total profit
total_profit = df["profit"].sum()
print("Total Profit:", total_profit)


# Highest selling category
top_category = df.groupby("category")["sales"].sum().sort_values(ascending=False)
print("\nTop Categories by Sales")
print(top_category)


# Average discount by category
avg_discount = df.groupby("category")["discount"].mean()
print("\nAverage Discount by Category")
print(avg_discount)
```

## Grouping

```python
# Sales by region
sales_by_region = df.groupby("region")["sales"].sum()

print("\nSales by Region")
print(sales_by_region)


# Profit by category
profit_by_category = df.groupby("category")["profit"].sum()

print("\nProfit by Category")
print(profit_by_category)


# Monthly sales
monthly_sales = df.groupby("order_month")["sales"].sum()

print("\nMonthly Sales")
print(monthly_sales)


# Region and category analysis
region_category = df.groupby(["region", "category"])["sales"].sum()

print("\nRegion Category Sales")
print(region_category)


# Sub-category profit analysis
subcategory_profit = df.groupby("sub-category")["profit"].mean()

print("\nAverage Profit by Sub-Category")
print(subcategory_profit)
```

# Save Cleaned Data

```python
df.to_csv("cleaned_superstore_dataset.csv", index=False)

print("\nDataset Saved Successfully")
```
